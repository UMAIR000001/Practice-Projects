```python
# Project_number ---02 ----- By Mohammad Umair Hafeez
''' 
This program is very useful because it estimates body fat percentage 
and health risk based on height and weight.
It can help identify potential health issues early so you can take action.
'''

# Formula: BMI = (weight in pounds * 703) / (height in inches * height in inches)

# Asking user if they want to calculate BMI using (kg/cm) or (pound/inches)
while True:  # Loop until the user provides valid input
    check_option = input("Do you want to find your Body Mass Index (BMI) using height in cm and weight in kg? [yes/no]: ").strip().lower()

    if check_option == 'yes':
        weight_kg = float(input("Enter your weight in kg: "))
        height_cm = float(input("Enter your height in cm: "))

        # Converting weight to pounds and height to inches
        weight = round(weight_kg * 2.2, 2)
        height = round(height_cm * 0.393701, 2)

        print(f"Converted Weight: {weight} lbs")
        print(f"Converted Height: {height} inches")
        break  # Exit the loop after valid input

    elif check_option == 'no':
        # Asking user for weight and height in pounds and inches
        weight = float(input("Enter your weight in pounds: "))
        height = float(input("Enter your height in inches: "))
        break  # Exit the loop after valid input

    else:
        print("Invalid input! Please enter 'yes' or 'no'.")

# BMI Calculation
bmi = round((weight * 703) / (height * height), 2)

# Identifying Health Condition Based on BMI
if bmi < 18.5:
    print(f"{bmi}  You are underweight")
    print("Health Risk: Minimal")
elif 18.5 <= bmi <= 24.9:
    print(f"{bmi}  You have a normal weight")
    print("Health Risk: Minimal")
elif 25 <= bmi <= 29.9:
    print(f"{bmi}  You are overweight")
    print("Health Risk: Increased")
elif 30 <= bmi <= 34.9:
    print(f"{bmi}  You are Obese")
    print("Health Risk: High")
elif 35 <= bmi <= 39.9:
    print(f"{bmi}  You are Severely Obese")
    print("Health Risk: Very High")
elif bmi >= 40:
    print(f"{bmi}  You are Morbidly Obese")
    print("Health Risk: Extremely High")
else:
    print("Invalid Input")
```
